**Q6) Create a distribution plot for the depth column and the normalized depth column. If there are negative values, ignore them.[2]**

```python
import scipy.stats as sp
import seaborn as sns

sns.distplot(df['Depth'])
```

```python
normalised_data = sp.boxcox(x=df[df['Depth']>0]['Depth'])
```

```python
fig,ax=plt.subplots(1,2)
sns.distplot(df['Depth'],ax=ax[0])
ax[0].set_title('Original Data')
sns.distplot(normalised_data[0],ax=ax[1])
ax[1].set_title('Normalized Data')
```

```python
normalised_data = sp.yeojohnson(x=df['Depth'])

fig,ax=plt.subplots(1,2)
sns.distplot(df['Depth'],ax=ax[0])
ax[0].set_title('Original Data')
sns.distplot(normalised_data[0],ax=ax[1])
ax[1].set_title('Normalized Data')
```

**Q7) Fill the missing values in the Magnitude type column based on the Magnitude column.[2]**

**Q8) Fill the missing values in the Root Mean Square column.[2]**


```python
df['Root Mean Square'].isnull().sum()
```

```python
df['Root Mean Square']=df['Root Mean Square'].fillna(df['Root Mean Square'].median())
```

```python
df['Root Mean Square'].isnull().sum()
```
