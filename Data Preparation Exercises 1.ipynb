**Q7) How many columns are dropped if we decide to eliminate columns that contain more than 1% of null values?**

```python
df.dropna(axis=1, thresh=1435)
```

**Q8) How many rows are dropped if we decide to eliminate rows that contain more than 5% of null values?**

```python
df.dropna(thresh=15)
```
