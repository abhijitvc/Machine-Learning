```python
df = pd.DataFrame({'student_name': ['Anastasia','Dima', 'Kathrine' ,'James','Emily','Michael','Matthew','Laura','Kevin','Jonas'],
                  'attempts' : [1,3,2,3,2,3,1,1,2,1],
                  'score' : [12.5,9,16.5,15.5,9,20,14.5,18.5,8,19],
                  'qualify': ['yes','no','yes','no','no','yes','yes','no','no','yes']})
```

```python
df
```

```python
#Select the rows where number of attempts in the examination is less than 3 and score greater than 15.
df[(df['attempts']<3) & (df['score']>15)]
```

```python
import pandas as pd

df3=pd.DataFrame({'loc_id':[1700,1800,2400,1900],
                  'city':['Mumbai','Texas','London','Delhi'],
                  'country_id':[101,204,305,101]
                })
df3
```
