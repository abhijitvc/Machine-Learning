```python
ani_fea=ani_df.loc[0:,'hair':'type']
```

```python
from sklearn import decomposition
from sklearn.decomposition import PCA
```

```python
pca=decomposition.PCA()
ani_pca=pca.fit_transform(ani_fea)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
print(pca.n_components_)
```

```python
pca=decomposition.PCA()
ani_pca=pca.fit_transform(ani_fea.loc[0:,'hair':'milk'])
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
print(pca.n_components_)
```

```python
pca=decomposition.PCA()
ani_pca=pca.fit_transform(ani_fea.loc[0:,'hair':'eggs'])
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
print(pca.n_components_)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(ani_fea)
scaled_data=scaler.transform(ani_fea)
pca=PCA(n_components=3)
ani_pca=pca.fit_transform(scaled_data)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
print(pca.n_components_)
```
